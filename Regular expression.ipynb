# code to match and their syntax

# Anchor (This are differnet)

```python
pattern = re.compile(r'\d\d\d') #matching every digit we have in our text_to _search
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'\d\d\d.\d\d\d.\d\d\d\d') #it will match our both phone number
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

# Qualifiers

```python
pattern = re.compile(r'\d{3}.\d{3}.\d{4}') 
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```
